# Hand Gesture Recognition with Convolutional Neural Networks (CNN)

This notebook demonstrates the process of building a Hand Gesture Recognition model using Convolutional Neural Networks (CNNs). The model is trained on the LeapGestRecog dataset, which consists of hand gesture images captured through Leap Motion sensors. The notebook covers data loading, preprocessing, model architecture definition, training, evaluation, and includes visualizations of training and validation accuracy/loss curves. The TensorFlow and Keras libraries are utilized for implementing the CNN, and the notebook serves as a practical guide for developing a gesture recognition system.

```python
# Continue with the rest of the code
import tensorflow as tf
from tensorflow.keras import layers, models
from sklearn.model_selection import train_test_split
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
from tqdm import tqdm
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.preprocessing.image import load_img, img_to_array
```

```python
# Load the dataset
dataset_path = "/content/leapgestrecog/leapGestRecog/00"
gestures = os.listdir(dataset_path)
```

```python
# Create a DataFrame to store file paths and labels
data = {'file_path': [], 'label': []}
```

```python
for gesture in gestures:
    gesture_path = os.path.join(dataset_path, gesture)

    if os.path.isdir(gesture_path):  # Check if it's a directory
        files = os.listdir(gesture_path)

        for file in tqdm(files, desc=gesture):
            file_path = os.path.join(gesture_path, file)
            data['file_path'].append(file_path)
            data['label'].append(gesture)
```

```python
# Convert the DataFrame to a Pandas DataFrame
df = pd.DataFrame(data)
```

```python
# Convert labels to integers using LabelEncoder
label_encoder = LabelEncoder()
df['label'] = label_encoder.fit_transform(df['label'])
```

```python
# Load and preprocess images
X = [img_to_array(load_img(file_path, target_size=(64, 64))) for file_path in df['file_path']]
X = np.array(X) / 255.0  # Normalize pixel values to [0, 1]
```

```python
# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, df['label'].values, test_size=0.2, random_state=42)
```

```python
# Define the model architecture
model = models.Sequential([
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(64, 64, 3)),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(128, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dense(len(set(df['label'])), activation='softmax')
])
```

```python
# Compile the model
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
# Train the model
history = model.fit(X_train, y_train, epochs=10, validation_data=(X_test, y_test))
```

```python
# Visualize training history
def plot_history(history):
    plt.figure(figsize=(12, 6))

    # Plot training & validation accuracy values
    plt.subplot(1, 2, 1)
    plt.plot(history.history['accuracy'])
    plt.plot(history.history['val_accuracy'])
    plt.title('Model accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend(['Train', 'Test'], loc='upper left')

    # Plot training & validation loss values
    plt.subplot(1, 2, 2)
    plt.plot(history.history['loss'])
    plt.plot(history.history['val_loss'])
    plt.title('Model loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend(['Train', 'Test'], loc='upper left')

    plt.tight_layout()
    plt.show()
```

```python
# Plot the training history
plot_history(history)
```
